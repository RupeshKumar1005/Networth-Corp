# Taking info from Product performance and updating it into Product description according to the respective product ID.

```python
for i in PP.columns[1:5]:
    PD[i]=None
```

```python
for i in PP.columns[1:5]:
    for j in PD['PID']:
        PD.loc[PD['PID']==j,i]=PP.loc[PP['pId']==j,i].values[0]
```

```python
PD.drop(columns=['pName'],inplace=True)
```

```python
PD
```

# changing type of clicks,carts and orders columns to integer type.

```python
for i in ['clicks','carts','orders']:
    PD[i]=PD[i].astype('int32')
```
